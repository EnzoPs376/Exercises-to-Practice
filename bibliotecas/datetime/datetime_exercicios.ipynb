## Oferecendo desconto para cliente com base na última compra

Suponha que você está trabalhando para uma empresa que deseja rastrear a atividade do cliente. Uma métrica que eles estão interessados é o tempo que passou desde a última transação do cliente. Se for muito tempo, eles podem oferecer um desconto para o cliente. Crie um script Python que mostra quanto tempo se passou desde a última compra do cliente (10 de abril de 2026). Se faz mais de 30 dias, mostre uma mensagem oferecendo um desconto para o cliente.

```python
from datetime import datetime
ultima_compra = datetime(2026, 4, 10)
data_hora_atual = datetime.now()
diferenca = data_hora_atual - ultima_compra
print(diferenca.days)
```

```python
if diferenca.days > 30:
    print("Você ganhou um desconto")
else:
    print("Você não ganhou um desconto")
```

## Data e hora em diferentes fusos horários

Uma empresa tem escritórios em São Paulo, Nova York e Tóquio. Crie um script Python que mostra a data e hora atuais nesses três fusos horários. Exiba, também, se estes escritórios estão abertos ou fechados (9h às 17h).


```python
from datetime import datetime
from zoneinfo import ZoneInfo

fuso_sp = ZoneInfo("America/Sao_Paulo")
fuso_ny = ZoneInfo("America/New_York")
fuso_to = ZoneInfo("Asia/Tokyo")

hora_atual = datetime.now()
hora_atual_sp = hora_atual.astimezone(fuso_sp)
hora_atual_ny = hora_atual.astimezone(fuso_ny)
hora_atual_to = hora_atual.astimezone(fuso_to)

print(f"Data/hora em São Paulo: {hora_atual_sp}")
print(f"Data/hora em Nova York: {hora_atual_ny}")
print(f"Data/hora em Toquio: {hora_atual_to}")
```

```python
def verifica_horario(data_hora):
    if 9 <= data_hora.hour < 17:
        return "aberto"
    else:
        return "fechado"

print(f"Escritório de São Paulo está {verifica_horario(hora_atual_sp)}")
print(f"Escritório de Nova York está {verifica_horario(hora_atual_ny)}")
print(f"Escritório de Toquio está {verifica_horario(hora_atual_to)}")
```
